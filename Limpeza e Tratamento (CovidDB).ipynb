## 1. Importacoes e Carga:

```python
# Importando Bibliotecas
import pandas as pd
import sqlite3 as sq3
```

```python
# Carregando os arquivos CSV em Dataframes
DF_Mortes = pd.read_csv("Dataset/covid_mortes.csv")
DF_Vacinas = pd.read_csv("Dataset/covid_vacinacao.csv")
```

### -  1.1 Criacao do Banco e Carga de Dados

```python
# Crianco Banco de Dados 
CovidDB = sq3.connect("Database/Covid.db")
```

```python
# Carregando 2 DataFrames em tabelas diferentes
DF_Mortes.to_sql("Mortes",CovidDB)
DF_Vacinas.to_sql("Vacinas",CovidDB)
```

### -  1.2 Consultas ao Banco

```python
# Carregando a extensao "SQl" para permitir consultas ao Banco de Dados
%load_ext sql
```

```python
# Montando conexao com o Banco criado
%sql sqlite:///Database/Covid.db
```
