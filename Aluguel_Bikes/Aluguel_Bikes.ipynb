```python
# Imports
import pandas as pd
import sqlite3
```

```python
# Criando Dataframe
DF = pd.read_csv("Dataset/tripdata_2022.csv")
DF.head()
```

# Metadados "tripdata_2023.csv"
- Duration – Duração da viagem (em segundos)
- Start Date – Data/Hora do inicio da viagem
- End Date –  Data/Hora do fim da viagem
- Start Station – Nome/Numero da Estação inicial
- End Station – Nome/Numero da Estação final
- Bike Number – ID da Bike usada na viagem
- Member Type – Indica se o usuario era registrado ("registered") ou casual ("casual")

```python
# Criando Banco de Dados
Bikesdb = sqlite3.connect("Database/Bikes.db")
```

```python
# Extraindo dados do Dataframe para a tabela alugueis
DF.to_sql("Alugueis",Bikesdb)
```

```python
# Carregando extensao SQL já que será usando a linguagem SQL nesse Notebook
%load_ext sql

# Montando conexão com o Banco de Dados Bikesdb
%sql sqlite:///Database/Bikes.db
```

```python
%%sql

SELECT * FROM Alugueis
LIMIT 10;
```

###### - Média de Tempo de duração do aluguel por tipo de membro:

```python
%%sql

SELECT MEMBER_TYPE "Tipo de Membro",
ROUND(AVG(DURATION)/60) "Duracao Media (minutos)"

FROM ALUGUEIS

GROUP BY MEMBER_TYPE;
```

###### - Média de Tempo de duração do aluguel por tipo de membro e por estação fim:

```python
%%sql

SELECT End_station "Estacao Final",
MEMBER_TYPE "Tipo de Membro",
ROUND(AVG(DURATION)/60) "Duracao Media (minutos)"

FROM ALUGUEIS

GROUP BY MEMBER_TYPE,
End_station
ORDER BY MEMBER_TYPE,
End_station;
```
